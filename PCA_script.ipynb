```python
# PCA without standardization
# 3 components
pca = PCA(n_components=3)

pcs = pca.fit_transform(df)
df['PC1'] = pcs[:,0]
df['PC2'] = pcs[:,1]
df['PC3'] = pcs[:,2]
```

```python
pca.explained_variance_ratio_
```

```python
# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df.select_dtypes(include=['float64', 'int64']))

# 3. Check explained variance
explained_variance_first_two = pca.explained_variance_ratio_.sum()

print(f"Variance explained by the first two components: {explained_variance_first_two*100:.2f}%")

if explained_variance_first_two >= 0.8:
    print("The first two components explain at least 80% of the variance!")
else:
    print("The first two components do not explain 80% of the variance.")
```
